```python
def movie_votes(vote):
    if vote in range(1000):
        return 1
    elif vote in range(1000,10000):
        return 2
    elif vote in range(10000,100000):
        return 3
    elif vote in range(100000,1000000):
        return 4
    else:
        return 5
```

```python
#CHALLENGE 4:
#To know if the sum of the ASCII value of every character of the movie title divided by the number of votes retrieve a prime number...
#remember, prime numbers are integers
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, num):
        if num % i == 0:
            return False
    return True    
```

```python
new_archivo['num_prime'] = new_archivo.apply(lambda x: is_prime(len(x['Title']) / x['Votes']), axis=1)
```
